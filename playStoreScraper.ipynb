```python
import pandas as pd
import numpy as np
from google_play_scraper import app, reviews_all, Sort
import plotly.express as px
from transformers import pipeline
```

```python
project = reviews_all('uk.gov.tfl.oystercontactless', sleep_milliseconds=0, lang = 'en', country = 'IN', sort=Sort.NEWEST)
```

```python
#project
```

```python
df = pd.json_normalize(project)
```

```python
df.head()
```

```python
df['score'].mean()
```

```python
df['reviewCreatedVersion'].value_counts()
```

```python
sentiment_analysis = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")
```

```python
print(sentiment_analysis("I love this!"))
```

```python
df.dtypes
```

```python
df['content'] = df['content'].astype('str')
```

```python
df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
df.head()
```

```python
df.shape
```

```python
df['sentiment'] = df['result'].apply(lambda x: (x[0]['label']))
df['score'] = df['result'].apply(lambda x: (x[0]['score']))
```

```python
df.head()
```

```python
df['score'].mean()
```

```python
df['sentiment'].value_counts(normalize=True)
```

```python
fig = px.histogram(df, x='sentiment', color='sentiment', title='Sentiment Analysis', text_auto=True)
fig.show()
```
